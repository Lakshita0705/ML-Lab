```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("BostonHousing.csv")
print("Missing values (True/False):")
print(df.isnull())

print("\nTotal missing values in each column:")
print(df.isnull().sum())
```

```python
before = df.shape
df_cleaned = df.dropna()
after = df_cleaned.shape

print("Dataset size before cleaning:", before)
print("Dataset size after cleaning:", after)
```

```python
df['rm'] = df['rm'].fillna(df['rm'].mean())
df['lstat'] = df['lstat'].fillna(df['lstat'].median())
print(df.isnull().sum())
```

```python
print("Before conversion:")
print(df.dtypes)

df['chas'] = df['chas'].astype(int)
df['rad'] = df['rad'].astype(int)

print("\nAfter conversion:")
print(df.dtypes)
```

```python
df = df.rename(columns={
    'crim': 'Crime Rate',
    'rm': 'Avg Rooms',
    'medv': 'House Price'
})

print(df.head())
```

```python
df['chas'] = df['chas'].replace({-1: 0, 2: 1})
print(df['chas'].unique())
```

```python
import matplotlib.pyplot as plt
plt.plot(df.index, df['House Price'])
plt.xlabel("Index")
plt.ylabel("House Price")
plt.title("House Price Trend")
plt.show()
df['chas'].value_counts().plot(kind='bar')
plt.xlabel("chas")
plt.ylabel("Count")
plt.title("CHAS Distribution")
plt.show()
```

```python
plt.scatter(df['Avg Rooms'], df['House Price'])
plt.xlabel("Average Rooms")
plt.ylabel("House Price")
plt.title("Rooms vs House Price")
plt.show()
```

```python
plt.hist(df['House Price'], bins=10)
plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.title("House Price Distribution")
plt.show()
```

```python
plt.boxplot(df['House Price'])
plt.xlabel("House Price")
plt.title("House Price Boxplot")
plt.show()
```
