```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"
df = pd.read_csv(url)
print(df.head())
print("\nNumerical Columns:")
print(df.select_dtypes(include=np.number).columns)
print("\nCategorical Columns:")
print(df.select_dtypes(include='object').columns)
```

```python
print("\nMean:")
print(df[['math score', 'reading score', 'writing score']].mean())

print("\nMedian:")
print(df[['math score', 'reading score', 'writing score']].median())

print("\nMode:")
print(df[['math score', 'reading score', 'writing score']].mode())
```

```python
print("\nMinimum Values:")
print(df[['math score', 'reading score', 'writing score']].min())

print("\nMaximum Values:")
print(df[['math score', 'reading score', 'writing score']].max())

print("\nSum:")
print(df[['math score', 'reading score', 'writing score']].sum())

print("\nVariance:")
print(df[['math score', 'reading score', 'writing score']].var())

print("\nStandard Deviation:")
print(df[['math score', 'reading score', 'writing score']].std())
```

```python
Q1 = df[['math score', 'reading score', 'writing score']].quantile(0.25)
Q2 = df[['math score', 'reading score', 'writing score']].quantile(0.50)
Q3 = df[['math score', 'reading score', 'writing score']].quantile(0.75)

print("First Quartile (Q1):\n", Q1)
print("\nSecond Quartile (Median - Q2):\n", Q2)
print("\nThird Quartile (Q3):\n", Q3)
```

```python
# Correlation matrix
correlation = df[['math score', 'reading score', 'writing score']].corr()
print("Correlation Matrix:\n", correlation)

# Covariance matrix
covariance = df[['math score', 'reading score', 'writing score']].cov()
print("\nCovariance Matrix:\n", covariance)
```

```python
df[['math score', 'reading score', 'writing score']].hist(bins=10)
plt.suptitle("Score Distributions")
plt.show()
```

```python
plt.boxplot([df['math score'], df['reading score'], df['writing score']],
            labels=['Math', 'Reading', 'Writing'])
plt.title("Boxplot of Student Scores")
plt.ylabel("Scores")
plt.show()
```
